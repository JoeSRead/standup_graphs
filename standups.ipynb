## New standup group

To create new standup groups: 
1) Load pickled graph G

2) If you need confirmation run standups(G) so you can get the groups approved 

3) Once confirmed update the graph by looping update(G, standup_group) over the list of groups

4) Deleate the old graph and pickle the new one

If you do not need approval all of the above can be run in a single step using stand_update(G) which will return the standup groups and update the graph (still need to pickle though)
