```python
def grouper(G, lm, assigned_lst=[], num =4):
    neb_lst = [lm]

    neb_weight = []

    for i in list(nx.neighbors(G, lm)):
        neb_weight.append((i, G[lm][i]['weight']))
    
    neb_weight = sample(neb_weight, len(neb_weight))
    neb_weight.sort(key = lambda x: x[1])

    for i in neb_weight:
        if(len(neb_lst) < num):
            if i[0] not in assigned_lst:
                neb_lst.append(i[0])
        else:
            break
    return neb_lst
```

```python
def line_managers(G, num=3):
    nn = [(node, len(list(nx.neighbors(G, node)))) for node in G.nodes]
    nn.sort(key= lambda x: x[1])
    return [n[0] for n in nn[:num]]
```

```python
line_managers(G)
```

```python
def standups(G, lms=[], nm =4):
    
    if len(lms) == 0:
        lm_lst = line_managers(G)
    else:
        lm_lst = lms
    
    groups_lst =[]
    as_lst = []
    
    if len(G.nodes) > len(lm_lst)*nm:
        
        print('Error: groups not large enough to cover entire team')
        return
    
    for lm in lm_lst:
        groups_lst.append(grouper(G, lm, as_lst, num = nm))
        as_lst.extend(grouper(G, lm, as_lst, num = nm))
    
    for mentor in list(G.nodes):
        if mentor not in as_lst:
            
            lm = [x for x in list(set(G.nodes) - set(G.neighbors(mentor))) if x != mentor][0]
            lm_neigh = [(i, G[lm][i]['weight']) for i in list(nx.neighbors(G, lm))]
            lm_neigh.sort(key = lambda x: x[1]) 
            
            for group in groups_lst:
                if lm_neigh[0][0] in group:
                    group.append(mentor)
                    group.remove(lm_neigh[0][0])
                if lm in group:
                    group.append(lm_neigh[0][0])
            
    return groups_lst
```

```python
standups(G, nm=4)
```

```python
def standup_update(G, group_lst):
    G.add_weighted_edges_from([(i,j,G[i][j]['weight'] + 1) for i in group_lst for j in group_lst if ((i!=j) & (i in nx.neighbors(G, j)))])
```
